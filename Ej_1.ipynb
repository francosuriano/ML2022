```python
# Importamos las bibliotecas necesarias

import pandas as pd
import numpy as npa
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stt
from google.colab import files
```

```python
# Cargamos el dataset de casos de covid

dataset = 'https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/4_DataWrangling/data/casos_covid_bahia.csv'
df_casos = pd.read_csv(dataset)
features = df_casos.columns.values
```

```python
# Observamos los features del dataset

print(df_casos.columns.values)
```

```python
# Función que muestra todos los histogramas del dataset

def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(20,10))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=20,ax=ax)
        ax.set_title(var_name)
    plt.show()

# descartamos el feature fecha

draw_histograms(df_casos[features[1:11]],features[1:11],3,4)

```


Una vez graficados los histogramas, procedemos a observar las distribuciones de los datos (si poseen curva normal, si se hallan muy sesgados, etc):
